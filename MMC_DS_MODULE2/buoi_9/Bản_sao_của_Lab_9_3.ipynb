# Lab: Làm việc với dữ liệu Object trong Pandas

# Đọc dữ liệu từ file

```python
import pandas as pd

df_google_playstore = pd.read_csv("../Google_Playstore.csv")
```

```python
df_google_playstore.head()
```

```python
# Yêu cầu: thay đổi gmail.com ở Developer Email thành gmails.com
# df_google_playstore["Developer Email"] = df_google_playstore["Developer Email"].replace(
#     to_replace="gmail.com", value="gmails.com"
# )
# TODO: Why use regex here? If using normal replace it won't work
df_google_playstore["Developer Email"].replace(
    to_replace=r"gmail.com$", value="gmails.com", regex=True, inplace=True
)
df_google_playstore["Developer Email"]
```

```python
# Chuyển đổi M trong cột Size thành MB, k trong cột Size thành kB
df_google_playstore["Size"].replace(
    to_replace="M$", value="MB", regex=True, inplace=True
)
df_google_playstore["Size"].replace(
    to_replace="k$", value="kB", regex=True, inplace=True
)
df_google_playstore["Size"]
```

```python
# Yêu cầu: Lấy các giá trị trước dấu chấm "." đầu tiên trong App Id
q = df_google_playstore["App Id"].apply(lambda x: x.split(sep=".")[0])
q
```

```python
# Yêu cầu: Quan sát các giá trị duy nhất và số lượng của giá trị trước dấu chấm "." đầu tiên trong App Id
# df_google_playstore["App Id"].apply(lambda x: x.split(sep=".")[0])
print(q)
print(q.count())
```

```python
# Yêu cầu: Tạo một cột mới chứa đuôi của Email trong Developer Email
# Because sometimes the Email can be missing, and it is not justified yet to completely omit
# the N/A rows, thus we will use try except with attribute error in NaN
def tail_email(row: pd.Series):
    try:
        result = row.split(sep="@")[-1]
    except AttributeError:
        result = None
    return result


df_google_playstore["Email Tail"] = df_google_playstore["Developer Email"].apply(
    tail_email
)
df_google_playstore["Email Tail"]
```
